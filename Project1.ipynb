```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from datetime import datetime, date
```

```python
#########################  NIKE ##############################################

df = pd.read_csv('NKE.csv',parse_dates=True)
df["Index"] = df.index
df['Volume']=df['Volume'].astype(float)
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.linear_model import ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
X = df[['Index','Close','Open','Volume']]
Y = df['Adj Close']
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.2,random_state = 10,shuffle=False)
scaler = StandardScaler().fit(X_train)

#Linear Regression

lm = LinearRegression()
lm.fit(X_train,Y_train)
predicted1 = lm.predict(X_test)
fig = plt.figure(figsize =(24,12))
plt.plot(X_test.index,Y_test)
plt.plot(X_test.index,predicted1)
plt.legend(["Actual Stocks","Predicted Stocks"],loc ="lower right")
plt.show()

#Decision Tree

dtr = DecisionTreeRegressor(random_state=0)
dtr.fit(X_train, Y_train)
predicted4 = dtr.predict(X_test)
fig = plt.figure(figsize =(24,12))
plt.plot(X_test.index,Y_test)
plt.plot(X_test.index,predicted4)
plt.legend(["Actual Stocks","Predicted Stocks"],loc ="lower right")
plt.show()

#Kneighbours

neigh = KNeighborsRegressor(n_neighbors=2)
neigh.fit(X_train, Y_train)
predicted3 = neigh.predict(X_test)
fig = plt.figure(figsize =(24,12))
plt.plot(X_test.index,Y_test)
plt.plot(X_test.index,predicted3)
plt.legend(["Actual Stocks","Predicted Stocks"],loc ="lower right")
plt.show()

#SVR

svr = SVR(kernel = 'rbf')
svr.fit(X_train, Y_train)
predicted5 = svr.predict(X_test)
fig = plt.figure(figsize =(24,12))
plt.plot(X_test.index,Y_test)
plt.plot(X_test.index,predicted5)
plt.legend(["Actual Stocks","Predicted Stocks"],loc ="lower right")
plt.show()
```
